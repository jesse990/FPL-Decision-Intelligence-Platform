```python
import pandas as pd
```

```python

files = [
    "C:/Users/JesseOnu/Downloads/DatasetsDataAnalystProjects/data_2023-24_player_idlist.csv_20250909_012700",
    "C:/Users/JesseOnu/Downloads/DatasetsDataAnalystProjects/data_2025-26_player_idlist.csv_20250909_012701",
    "C:/Users/JesseOnu/Downloads/DatasetsDataAnalystProjects/data_2024-25_player_idlist.csv_20250909_012701"
]

# 2. Load each into a DataFrame
dfs = [pd.read_csv(f) for f in files]

# 3. Concatenate into one DataFrame
df = pd.concat(dfs, ignore_index=True)

# 4. Create FullName column
df["FullName"] = df["first_name"] + " " + df["second_name"]

# 5. Remove duplicates by FullName
df = df.drop_duplicates(subset=["FullName"])

# 6. Drop old ID column
df = df.drop(columns=["id"], errors="ignore")

# 7. Create new unique ID
df["PlayerID"] = df["FullName"].astype("category").cat.codes

# Save
#df.to_csv("players_master.csv", index=False)
```
