```python
# function for adding gameweek column and concatinating

def load_gw_files(folder, prefix, output_file=None):
    dfs=[]
    for file in os.listdir(folder):
        if file.startswith(prefix):
            match = re.search(rf'{prefix}(\d+)', file)
            gameweek = int(match.group(1)) if match else None

            df = pd.read_csv(os.path.join(folder, file))
            df["GameWeek"] = gameweek
            dfs.append(df)
    final_df = pd.concat(dfs, ignore_index=True)
    if output_file:
        final_df.to_csv(output_file, index = False)
    return final_df
```
